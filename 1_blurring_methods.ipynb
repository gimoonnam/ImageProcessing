```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../images/lena.png')
#plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))

def plotFigures(original_image, blurred_image):
    plt.figure(figsize=(10,10))
    plt.subplot(121)
    plt.imshow(cv2.cvtColor(original_image,cv2.COLOR_BGR2RGB))
    plt.title('Original')
    plt.xticks([]), plt.yticks([])

    plt.subplot(122)
    plt.imshow(cv2.cvtColor(blurred_image,cv2.COLOR_BGR2RGB))
    plt.title('Blurred')
    plt.xticks([]), plt.yticks([])
    plt.show()
```

## 1. Averaging blurring

```python
blur1 = cv2.blur(image,(5,5))
plotFigures(image,blur1)
```

## 2. Gaussian blurring
depth: CV_8U,CV_16U,CV_16S,CV_32F,CV_64F
$$$$
cv2.GaussianBlur(src_image,(ksize_x,ksize_y),sigmaX)
- kernel size in x and y: these can differ 
- sigmaX: STD of gaussian kernel. If sigmaX is set to 0, then sigmaY is also automatically set to 0


```python
blur2 = cv2.GaussianBlur(image,(15,15),0)
plotFigures(image,blur2)
```

## 3. Median blurring

```python
median = cv2.medianBlur(image,7)
plotFigures(image,median)
```

## 4. Bilateral filtering 

highly effective in noise removal while keeping edges sharp. but the operation is slow compared to other filters. 

Bilateral filtering also takes a Gaussian filter in space, but what's more than Gaussian filter is a function of pixel difference. So the Bilateral filter considers the pixels with the similar intensities to the centeral pixel for blurring, allowing to preserve the edges since pixels at edges will have large intensity variation. 
$$ $$
cv2.bilateralFilter(src_image,d,sigmaColor,sigmaSpace)
d: diameter around central pixel for filttering 

sigmaColor: filter sigma in the color space: A larger parameter means that farther pixels will influence each other as long as their colors are close enough 

sigmaSpace: Filter sigma in the coordinate space. A larger value of the parameter means that farther pixels will influence each other as long as their colors are close enough (see sigmaColor ). When d>0, it specifies the neighborhood size regardless of sigmaSpace. Otherwise, d is proportional to sigmaSpace.

```python
bilateral = cv2.bilateralFilter(image,10,75,75)
plotFigures(image,bilateral)
```
